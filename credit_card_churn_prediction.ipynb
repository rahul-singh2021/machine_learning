```python
import pandas as pd
data=pd.read_csv("/content/Churn_Modelling[1].csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(columns =['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
data.head()
```

```python
data['Geography'].value_counts()
```

```python
data['Gender'].value_counts()
```

```python
data=pd.get_dummies(data,columns=['Geography','Gender'],drop_first=True)
```

```python
data.head()
```

```python
data.columns
```

```python
x=data.drop(columns=['Exited'])
```

```python
y=data['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

x_train_trf=scaler.fit_transform(x_train)
x_test_trf=scaler.fit_transform(x_test)
```

```python
x_train_trf
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(7,activation='sigmoid',input_dim=11))
model.add(Dense(5,activation='sigmoid'))
model.add(Dense(4,activation='sigmoid'))
model.add(Dense(4,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy')
```

```python
history=model.fit(x_train,y_train,batch_size=50,epochs=1000)
```

```python
model.layers[1].get_weights()
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
import numpy as np
```

```python
final_pred=np.where(y_pred>=0.5,1,0)
```

```python
final_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(final_pred,y_test)
```
