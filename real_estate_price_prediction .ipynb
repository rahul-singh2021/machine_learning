```python
# import all prerequired libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(20,10)
```

```python
# load the bengaluru home prices data into dataframe
df1=pd.read_csv('Bengaluru_House_Data.csv')
df1.head()
```

```python
# number of rows and columns in our dataset
df1.shape
```

```python
# count the number of each category of area_type
df1.groupby('area_type')['area_type'].agg('count')
```

```python
# drop less important columns and save it in new dataframe
df2=df1.drop(['area_type','society','balcony','availability'],axis='columns')
df2.head()
```

```python
# data cleaning or filling NA values
df2.isnull().sum()
```

```python
# drop all rows with with NA values
df3=df2.dropna()
df3.isnull().sum()
```

```python
# applying feature engineering to add new feature(integer) for bhk
df3['bhk']=df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3.head()
```

```python
# exploring total_sqft feature
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
# conveting the range and non sqft values of total_sqft to a valid number
def convert_sqft_to_num(x):
  tokens=x.split('-')
  if len(tokens)==2:
    return (float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df4=df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(10)
```

```python
# adding new feature price per sqft which can help in oulier detection
df5=df4.copy()
df5['price_per_sqft']=df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
# categorising thre location column
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10=location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
# moving all location having location_stats less thrn 10
df5.location=df5.location.apply(lambda x:'others' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
 # removing outliers
 df5[df5.total_sqft/df5.bhk<300].head()
```

```python
df5.shape
```

```python
df6=df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
# outlier removal using standard deviation and mean
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
```

```python
# checking for same area and location does small property have high prices as compare to large property
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    plt.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df7,"Rajaji Nagar")
```

```python
plot_scatter_chart(df7,"Hebbal")
```

```python
# removing the bhk outliers
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_bhk_outliers(df7)
# df8 = df7.copy()
df8.shape
```

```python
# droping some unnecessary features
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head()
```

```python
# converting text column into numeric
dummies=pd.get_dummies(df10.location)
dummies.head()
```

```python

df11 =pd.concat([df10,dummies.drop('others',axis='columns')],axis='columns')
df11.head()
```

```python
df12=df11.drop(['location'],axis='columns')
df12.head()
```

```python
df12.shape
```

```python
x=df12.drop(['price'],axis='columns')
x.head()
```

```python
y=df12.price
y.head()
```

```python
# splitting out data for training and testing purpose
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
# Linear Regression model
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
# using K fold cross validation for accuracy of our model
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)

cross_val_score(LinearRegression(),x,y,cv=cv)
```

```python
# finding best model using grid cv
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'n_jobs': [True, False]
            }

        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)

```

```python
# defining a function to predict prices
def predict_price(location,sqft,bath,bhk,x):
    loc_index = np.where(x.columns==location)[0][0]

    x = np.zeros(len(x.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
